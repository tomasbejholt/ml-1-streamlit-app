## What is Machine Learning?

In traditional programming, you write rules: "if the email contains these words, it's spam." Machine learning flips that - instead of writing rules, you show the system thousands of examples of spam and not-spam, and it figures out the rules on its own. The rules it discovers are often subtler and more effective than anything a human would write by hand.

| Traditional Programming | Machine Learning |
|------------------------|-----------------|
| You write the rules | The algorithm discovers the rules |
| Rules + Data = Answers | Data + Answers = Rules |
| Breaks when patterns change | Retrain on new examples |
| Hard to scale to complex problems | Scales with data |

Here's what that looks like in practice. Say you want to predict whether an email is spam:

| Email text | Contains link? | ALL CAPS words | Known sender? | **Spam?** |
|-----------|:-:|:-:|:-:|:-:|
| "You won $1M! Click here" | Yes | 3 | No | **Yes** |
| "Meeting at 3pm tomorrow" | No | 0 | Yes | **No** |
| "URGENT: Verify account NOW" | Yes | 2 | No | **Yes** |
| "Here's the report you asked for" | Yes | 0 | Yes | **No** |
| "Congratulations! You've been selected" | Yes | 1 | No | **Yes** |
| "Lunch at noon?" | No | 0 | Yes | **No** |

A traditional programmer looks at this and writes rules ("if ALL CAPS > 1 and not from a known sender, it's spam"). An ML model looks at thousands of these rows and discovers its own rules - including patterns a human might miss, like specific combinations of link placement, word frequency, and sending time.

At its core, ML is about learning patterns from data. A model starts with random guesses, sees examples with correct answers, and gradually adjusts its internal numbers (called **weights**) until it gets good at predicting. That adjustment process is called **training**. Using the trained model on new data is called **inference**.

## Types of Machine Learning

We talk a lot about supervised and unsupervised learning when talking about ML. Supervised learning the most common one, and unsupervised learning is this weird collection of algorithms that can be useful sometimes. Think of these areas of machine learning as algorithms or techniques that solve specific types of problems.

| Type | What it needs | What it does | Examples |
|------|:-------------|:------------|:---------|
| **Supervised** | Labeled data (input + correct answer) | Learns to predict the answer for new inputs | Spam detection, price prediction, image classification |
| **Unsupervised** | Unlabeled data | Finds structure and patterns on its own | Customer segmentation, anomaly detection |
| **Self-supervised** | Raw data (text, images) | Creates its own labels from the data | LLM pretraining ("predict the next word") |
| **Reinforcement** | Environment + rewards | Learns by trial and error | Game AI (AlphaGo), robotics |

Supervised learning is the focus of this course. You have labeled data - inputs paired with correct answers - and the model learns the mapping. It comes in two flavors: classification (predict a category) and regression (predict a number). This thing called "label" is super important. The label is the answer - it's the data we have history on, but in the future we might not have the label. Here's what the data actually looks like:

**Classification - "will this passenger survive?"**

The columns are the **features** (inputs). The last column is the **label** (what we predict).

| Name | Class | Sex | Age | Fare | Family onboard | **Survived (label)** |
|------|:-----:|-----|----:|-----:|:--------------:|:-------------------:|
| Mrs. John Bradley Cumings | 1st | female | 38 | $71.28 | 1 | **Yes** |
| Mr. Owen Harris Braund | 3rd | male | 22 | $7.25 | 1 | **No** |
| Master Gosta Leonard Palsson | 3rd | male | 2 | $21.08 | 4 | **No** |
| Miss Laina Heikkinen | 3rd | female | 26 | $7.93 | 0 | **Yes** |
| Mr. Timothy J McCarthy | 1st | male | 54 | $51.86 | 0 | **No** |

**Regression - "what will this house sell for?"**

| Size (sqft) | Bedrooms | Year built | Distance to city center | Garage | **Price (label)** |
|:-----------:|:--------:|:----------:|:----------------------:|:------:|:-----------------:|
| 1,400 | 3 | 1995 | 12 km | Yes | $285,000 |
| 850 | 1 | 2010 | 2 km | No | $340,000 |
| 2,200 | 4 | 1980 | 18 km | Yes | $410,000 |
| 650 | 1 | 2020 | 0.5 km | No | $520,000 |

Same idea, different output. Classification predicts a category. Regression predicts a number. The model's job is to find patterns in the features that help predict the label.

Self-supervised learning is worth understanding because it's how modern LLMs work. The model just predicts the next word, over and over, on massive amounts of text. No human labeling needed. We'll work with these models through APIs later in the course.

## Classic ML vs Deep Learning

For most of ML's history, the hard part was feature engineering - manually designing the inputs. You'd spend weeks crafting features like word counts, sender reputation scores, or pixel histograms. The model itself (logistic regression, random forest) was almost an afterthought.

Deep learning changed this for images and text. Instead of hand-crafting features, you feed raw pixels or words into a deep neural network and let it learn its own features. Early layers learn simple patterns (edges, common word pairs), later layers learn complex concepts (faces, sentence meaning).

| Data type | Best approach | Why | When in this course |
|-----------|--------------|-----|:-------------------:|
| Tabular (spreadsheets, databases) | Classic ML - trees, XGBoost | Feature engineering helps, limited data | L3-L7 |
| Images (photos, scans) | Deep learning - CNNs | Learns features from raw pixels | L9-L12 |
| Text (documents, reviews) | Deep learning - Transformers, LLMs | Sequential structure, massive pretraining | L13+ |

We'll learn both approaches. Trees still dominate tabular data, and that's totally fine. The right tool depends on the problem.

### Demo 1: Predicting Titanic Survival (Tabular Data)

The Titanic sank on April 15, 1912. Over 1,500 of the ~2,200 people aboard died. But survival wasn't random — it depended on factors like gender, class, and age ("women and children first" was real policy, and first-class passengers had better access to lifeboats).

Given a passenger's information, can we predict whether they survived? This is **binary classification**: two outcomes, survived or died.

````python
# Load and explore the data
titanic = pd.read_csv('../../../../data/titanic/titanic.csv')

print(f"Dataset: {len(titanic)} passengers, {len(titanic.columns)} columns")
print(f"Survived: {titanic['Survived'].sum()} ({titanic['Survived'].mean():.0%})  |  Died: {(~titanic['Survived'].astype(bool)).sum()} ({1-titanic['Survived'].mean():.0%})")
print()
titanic[['Name', 'Pclass', 'Sex', 'Age', 'Fare', 'Survived']].head(8)
````

Every row is a passenger. Every column is a **feature** — a piece of information the model can use. The **label** (what we're predicting) is the `Survived` column.

Before building any model, let's look for patterns in the data. Are some groups more likely to survive?

````python
fig, axes = plt.subplots(1, 3, figsize=(14, 4))

# By gender
surv_sex = titanic.groupby('Sex')['Survived'].mean()
axes[0].bar(['Female', 'Male'], [surv_sex['female'], surv_sex['male']], color=['#e74c3c', '#3498db'])
for i, v in enumerate([surv_sex['female'], surv_sex['male']]):
    axes[0].text(i, v + 0.02, f'{v:.0%}', ha='center', fontweight='bold')
axes[0].set_ylabel('Survival Rate')
axes[0].set_title('By Gender')
axes[0].set_ylim(0, 1)

# By class
surv_class = titanic.groupby('Pclass')['Survived'].mean()
axes[1].bar(['1st', '2nd', '3rd'], surv_class.values, color=['#2ecc71', '#f39c12', '#9b59b6'])
for i, v in enumerate(surv_class.values):
    axes[1].text(i, v + 0.02, f'{v:.0%}', ha='center', fontweight='bold')
axes[1].set_ylabel('Survival Rate')
axes[1].set_title('By Ticket Class')
axes[1].set_ylim(0, 1)

# Age distribution
axes[2].hist(titanic[titanic['Survived']==1]['Age'].dropna(), bins=20, alpha=0.7, label='Survived', color='#3498db')
axes[2].hist(titanic[titanic['Survived']==0]['Age'].dropna(), bins=20, alpha=0.7, label='Died', color='#e74c3c')
axes[2].set_xlabel('Age')
axes[2].set_ylabel('Count')
axes[2].set_title('Age Distribution')
axes[2].legend()

plt.tight_layout()
plt.show()
````

Clear patterns: women survived at 3x the rate of men, first class at nearly 3x third class, and children had better odds. A human could write rules for this ("if female and first class, predict survived"). But ML learns these rules automatically — and finds subtler patterns we'd miss.

````python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

# Prepare: select features, handle missing values, encode sex as numbers
df = titanic[['Pclass', 'Sex', 'Age', 'Fare', 'SibSp', 'Parch', 'Survived']].copy()
df['Sex'] = df['Sex'].map({'male': 1, 'female': 0})
df = df.dropna()

# Split: 80% for training, 20% for testing
X_train, X_test, y_train, y_test = train_test_split(
    df.drop('Survived', axis=1), df['Survived'], test_size=0.2, random_state=42
)

# Train a Random Forest (100 decision trees working together)
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

print(f"Training accuracy: {model.score(X_train, y_train):.0%}")
print(f"Test accuracy:     {model.score(X_test, y_test):.0%}")
print(f"\nThe model predicts survival correctly for ~{model.score(X_test, y_test):.0%} of passengers it's never seen.")
````

Six features, ten lines of code, 77% accuracy on unseen data. The model automatically discovered that sex and class matter most — the same patterns we spotted in the charts, plus subtler interactions we didn't.

The gap between training accuracy (99%) and test accuracy (77%) is called **overfitting** — the model memorized the training data a bit too well. We'll learn how to manage this throughout the course.

Key vocabulary from this demo: **features** (the input columns), **labels** (what we predict), **training** (fitting the model), **test set** (data held out for evaluation), **overfitting** (memorizing training data instead of learning general patterns).

### Demo 2: Recognizing Animals in Photos (Image Data)

Tabular data is rows and columns. Images are completely different — grids of pixel values. A 224×224 color photo is 224 × 224 × 3 = **150,528 numbers**. Red, green, and blue intensity for every pixel.

How do you go from 150,000 numbers to "this is a beagle"? That's what neural networks learn to do. Early layers detect edges, middle layers detect textures and shapes, later layers detect ears, eyes, snouts. The network builds up from simple patterns to complex concepts, entirely from examples.

````python
import torch
from torchvision import models, transforms
from torchvision.datasets import OxfordIIITPet
from PIL import Image
import os, urllib.request

# Download Oxford Pets if not already present
pets_dataset = OxfordIIITPet(root='./data', download=True)

# Show what the computer actually sees
sample_img, _ = pets_dataset[42]
img_array = np.array(sample_img)

fig, axes = plt.subplots(1, 4, figsize=(15, 3.5))
axes[0].imshow(sample_img); axes[0].set_title('What we see'); axes[0].axis('off')
axes[1].imshow(img_array[:,:,0], cmap='Reds'); axes[1].set_title('Red channel'); axes[1].axis('off')
axes[2].imshow(img_array[:,:,1], cmap='Greens'); axes[2].set_title('Green channel'); axes[2].axis('off')
axes[3].imshow(img_array[:,:,2], cmap='Blues'); axes[3].set_title('Blue channel'); axes[3].axis('off')
plt.suptitle(f'Image shape: {img_array.shape[1]}×{img_array.shape[0]} pixels × 3 channels = {img_array.size:,} numbers', fontsize=11)
plt.tight_layout()
plt.show()
````

Every pixel is a number between 0 (black) and 255 (white). Three channels — red, green, blue — stack to make color. The model's job is to find patterns in these numbers that distinguish a beagle from a bulldog.

Training an image classifier from scratch requires millions of images. Instead, we'll use **transfer learning**: take a model that someone already trained on 1.2 million images (ResNet-50, trained on ImageNet) and apply its knowledge to our photos. It already knows edges, textures, shapes, and hundreds of object categories.

````python
# Load a pretrained ResNet-50 (25 million parameters, trained on 1.2M images)
resnet = models.resnet50(weights=models.ResNet50_Weights.IMAGENET1K_V2)
resnet.eval()

# Load ImageNet class labels
labels_path = '/tmp/imagenet_classes.txt'
if not os.path.exists(labels_path):
    urllib.request.urlretrieve(
        'https://raw.githubusercontent.com/pytorch/hub/master/imagenet_classes.txt', labels_path)
with open(labels_path) as f:
    imagenet_labels = [l.strip() for l in f.readlines()]

# Preprocessing pipeline (must match what the model was trained on)
preprocess = transforms.Compose([
    transforms.Resize(256), transforms.CenterCrop(224), transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])

def classify(img):
    with torch.no_grad():
        probs = torch.nn.functional.softmax(resnet(preprocess(img).unsqueeze(0))[0], dim=0)
    top5 = torch.topk(probs, 5)
    return [(imagenet_labels[i], p.item()) for i, p in zip(top5.indices, top5.values)]

# Classify 5 random pet images
import random; random.seed(7)
pets_dir = './data/oxford-iiit-pet/images'
sample_files = random.sample([f for f in os.listdir(pets_dir) if f.endswith('.jpg')], 5)

fig, axes = plt.subplots(1, 5, figsize=(16, 3.5))
for ax, fname in zip(axes, sample_files):
    img = Image.open(os.path.join(pets_dir, fname)).convert('RGB')
    preds = classify(img)
    ax.imshow(img); ax.axis('off')
    ax.set_title(f'{preds[0][0]}\n{preds[0][1]:.0%}', fontsize=9)
plt.suptitle('ResNet-50 Predictions on Pet Photos', fontsize=12)
plt.tight_layout()
plt.show()
````

A model trained on general images (ImageNet) can recognize specific dog and cat breeds — even though it was never specifically trained on this pet dataset. That's the power of **transfer learning**: knowledge learned on one task transfers to related tasks.

### Demo 3: Understanding and Generating Text (NLP)

The most dramatic ML progress in recent years has been in language. Models based on the **Transformer** architecture can understand context, nuance, and intent in ways that seemed impossible a few years ago.

Two quick demos: sentiment analysis (classifying text) and text generation (creating text).

````python
from transformers import pipeline

# Sentiment analysis — classify movie reviews as positive or negative
sentiment = pipeline('sentiment-analysis', model='distilbert-base-uncased-finetuned-sst-2-english')

reviews = [
    "This movie was absolutely fantastic! Best film I've seen in years.",
    "Terrible waste of time. The plot made no sense.",
    "It was okay. Nothing special, but not bad either.",
    "The visuals were stunning but the story felt hollow and predictable.",
]

print("Sentiment Analysis")
print("=" * 60)
for review in reviews:
    r = sentiment(review)[0]
    icon = "+" if r['label'] == 'POSITIVE' else "-"
    short = review[:55] + ('...' if len(review) > 55 else '')
    print(f'  [{icon}] {r["label"]:8s} ({r["score"]:.0%})  "{short}"')

print()

# Text generation — a tiny local model (production LLMs are far more capable)
gen = pipeline('text-generation', model='distilgpt2', max_length=40, pad_token_id=50256)
result = gen("Machine learning is", num_return_sequences=1, do_sample=True, temperature=0.7)
print("Text Generation (distilgpt2 — a tiny model)")
print("=" * 60)
print(f'  Prompt: "Machine learning is"')
print(f"  Output: {result[0]['generated_text']}")
print()
print("This is a 82M parameter model. GPT-4 has ~1.8 trillion. Claude, LLaMA, Gemini")
print("are in the same league. We'll work with these APIs later in the course.")
````

The sentiment model correctly handles straightforward cases and even picks up mixed signals ("visuals were stunning but story felt hollow" → negative). The text generator... produces coherent-ish English but nothing insightful. That's because it's a tiny 82M parameter model. The LLMs you've used (ChatGPT, Claude) have 100-1000x more parameters trained on vastly more data.

The key insight: these models weren't programmed with grammar rules or sentiment dictionaries. They learned language structure from enormous amounts of text. The sentiment model was fine-tuned on labeled movie reviews. The generator was trained to predict the next word, billions of times. That's it.

## The ML Workflow

Every ML project — from a weekend experiment to a production system — follows the same loop:

```
1. UNDERSTAND     What are you predicting? What data do you have?
       ↓
2. PREPARE        Clean, transform, split into train/test
       ↓
3. TRAIN          Choose a model, fit it to training data
       ↓
4. EVALUATE       Test on held-out data — how good is it, really?
       ↓
5. ITERATE        Try different models, tune settings, improve features
       ↓
6. DEPLOY         Put it into production, monitor, retrain as needed
```

We followed steps 1-4 in the Titanic demo. In Lesson 2, we'll go through all six with an image classifier. By Lesson 7, this loop will feel automatic.

## Summary

**What ML is:** Learning patterns from data to make predictions. Data + Answers → Rules.

**Where it lives:** Recommendations, search, vision, language, fraud detection, science. Anywhere rules are hard to write but examples are plentiful.

**Types:** Supervised learning (our focus — classification and regression), unsupervised (clustering, anomaly detection), self-supervised (how LLMs learn).

**Classic vs modern:** Feature engineering + trees for tabular data. Deep learning + learned features for images and text. Both are current, both are useful, and we'll learn both.

**The workflow:** Understand → Prepare → Train → Evaluate → Iterate → Deploy. Every project, every lesson.

**Next lesson:** We train a real image classifier with fastai — from raw photos to a working model in minutes. You'll see the full workflow in action before understanding the internals.
